# Introduction/Business Problem


Motor vehicle travel is the main means of transportation in the world. It provide great convenience and opportunity for personal control and autonomy. However there are problems need to be addressed for improved safty.

This project will build a predicting system to identify the factors which are effecting traffic crash severity, predict the crach severity class based on selected factors such as weather condition, road condition, Speed and traffic control condition. 
Based on the result or the understanding, Transport department would have an idea or direction to prevent and reduce traffic accident, such as improving road condition, adding traffic light/signs; People would pay attention at known Accident prone area that and take carefully deal with various situations. Also for car verdors, those information might be a good guidance for improve car safety equipment/sensor.

Then we have the following business requirements as the purpose of project.
1. Understand explicitly what factors led to the predicted severity class
2. Predict Car Crash Severity class for each accident

According to the particular dataset(Data-Collisions.csv) which is provided in this course, there are only two crash severity - 1. Property Damage Only Collision and 2. Injury Collision, but as checked other data source(eg. Victoria Australia and New Zealand's data, the Severity are divided to four class), so I consider this problem belongs to multiple classification.
